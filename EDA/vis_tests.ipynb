```python
def players_join_ha(player_df):
    player_df = pd.DataFrame({'season_id': player_df['season_id'],
                              'player_id': player_df['player_id'],
                              'nmr_games': player_df['nmr_games'],
                              'fg_pts': player_df['fg_pts_home'] + player_df['fg_pts_away'],
                              'ast': player_df['ast_home'] + player_df['ast_away'],
                              'miss': player_df['miss'],
                              'blk': player_df['blk_home'] + player_df['blk_away'],
                              'ft_misses': player_df['ft_misses'],
                              'ft_scored': player_df['ft_home'] + player_df['ft_away'],
                              'reb': player_df['reb_home'] + player_df['reb_away'],
                              'tov': player_df['tov_home'] + player_df['tov_away'],
                              'stl': player_df['stl_home'] + player_df['stl_away'],
                              'pf': player_df['pf_home'] + player_df['pf_away']})
    return player_df
```

```python
players_join_ha(row)
```
